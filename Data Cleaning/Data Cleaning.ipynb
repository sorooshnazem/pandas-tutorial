# Non Standard Types

```python
df=pd.read_csv('data.csv',na_values=["'NaN'",'na'])
df
```

# Unexpected Types

```python
i=0
for age in df['age']:
    try:
        int(age)
    except:
        df.loc[i,'age']=np.nan
        pass
    i+=1
```

```python
df
```

```python
df.loc[(df['gender']!='m') & (df['gender']!='f'),'gender']=np.nan
```

```python
import pandas as pd
```

```python
data1={'name':['Soroosh','Mohammad','Maryam'], 'gender':['m','m','f']}
df1=pd.DataFrame(data1)
df1
```

```python
df1.index
```

```python
data2={'name':['Ali','Mahnaz','Saman'], 'gender':['m','f','m']}
df2=pd.DataFrame(data2)
df2
```

```python
df=df1.append(df2,ignore_index=True)
df
```

```python
df.index+=1
df
```

```python
df.index=df['name']
df
```

```python
df=df.set_index('name')
df
```

```python
df=df.reset_index()
df
```

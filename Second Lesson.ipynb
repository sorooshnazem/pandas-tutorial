```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airports=pd.read_csv('airports.csv')
```

```python
airports['iso_country'][pd.isnull(airports['continent'])].unique()
```

```python
airports.loc[pd.isnull(airports['continent']),'continent']='NA'
```

```python
airports['type'].value_counts().plot(kind='bar',figsize=(7,7))
plt.xlabel('Airports Types')
plt.ylabel('Count')
plt.show()
```

```python
airports['type'].value_counts().plot(kind='pie',figsize=(8,8))
plt.show()
```

```python
airports.groupby(['continent','type']).size()
```

```python
airports.groupby(['continent','type']).size().unstack().plot(kind='bar',stacked=True,figsize=(10,10))

plt.show()
```

```python
plt.savefig('airports_type_continent.png')
```
